# 정형 변환, 랜덤한 위치의 공 개수

## 신경망으로 홀수와 짝수를 구분하는 알고리즘을 생각중에 떠오른 아이디어

안녕하세요 강필성교수님 제자 송서하석사과정입니다.

갑작스럽게 연락드린 이유는 새로운 논문 아이디어

(물론 기존 진행중인 논문들도 여전히 유요합니다.)

가 떠올랐는데

이것이 논문감이 되는것인지 혼자서는 판단이 서지 않아 연락드렸습니다.

연구의 결론부터 말씀드리면...

"일반 신경망에서 활성함수만을 변경해 RNN과 동일한 신경망을 만들 수 있다는 이론"

입니다.

여기서 활성함수는

{시드모이드, 렐루, ...} => {sin, cos, ...}

위처럼 일반 활성함수를 파동함수로 변화시킵니다.

아이디어의 시작은 지인이 일반 신경망으로 홀수 짝수를 구분할수 있느냐는 질문에서 부터입니다.

저는 1차원 공간에서 대충 홀짝의 분포를 상상해 보았고,

이를 나누기 위해선 다음처럼 무한한 파동함수가 필요하다는 것을 깨달았습니다.

![](img/1.png)

경험적으로 생각해보았을 때 파동함수를 사칙연산으로 근사하기 위해서 테일러 근사를 하는데

무한한 영역에 파동함수를 테일러근사 하는 것은 무한개의 파라미터가 필요합니다.

![](img/2.png)

그러므로 분명 파동함수는 일반 신경망으로는 근사할수 없을것이라는 결론에 도달해

이 문제는 일반신경망으로는 풀수 없고 RNN으로 접근해야 한다고 결론을 냈습니다.

그러던중,

1. "RNN이 시퀀스에 대한 파동함수 역할을 한다."

2. "그럼 굳이 RNN으로 풀 필요가 없이 활성함수를 sin함수로 바꾸면 되는게 아닌가"

라는 생각이 떠올랐습니다.

이 문제는 신경망의 활성함수를 파동함수를 쓴다면 쉽게 일반 신경망으로도 풀리게 됩니다.
