# XML 전환 모듈

```python
def arrdic2xml(arrdic):
    
    def dic2tag(dic):
        tag = list(dic.keys())[0]
        value = (arrdic2xml(dic[tag])
                 if type(dic[tag]) == list
                 else "<![CDATA[{}]]>".format(dic[tag]))
        return "<{0}>{1}</{2}>".format(tag, value, tag.split()[0])
    
    return "".join([dic2tag(dic) for dic in arrdic])
```

```python
text = "테스트 텍스트 ㅎㅎㅎㅎ"

arrdic = [
    {"configure": [
        {"experiment name=\"키보드실험\"": [
            {"challenge type=\"keyboard\"": text}
        ]}
    ]}
]
```

```python
from bs4 import BeautifulSoup
result = arrdic2xml(arrdic)
bs = BeautifulSoup(result, "xml")
print(bs.prettify())
```

# 명언 Parser

```python
import re


with open("./speech-kr.txt") as f:
    text = f.read()

p = re.compile("(\d+\.)*(.*)*(\[.*)*")
lines = text.split("\n")
speechs = []
collector = ""
for line in lines:
    outputs = p.findall(line)[0]
    print(outputs)
    if outputs[0]:
        speech = " ".join(collector.split())
        speechs.append(speech)
        collector = ""
    collector += outputs[1]
speechs = set(speechs) - {""}
print(len(speechs))
speechs
```
