# Finding the quickest way to solve pattern quiz

안녕하세요 존경하는 교수님과 선배님들과 후배님들. 올해로 2학기를 지내고 있는 송서하 석사과정입니다.

이번 개인연구 발표를 준비하면서, 이번에는 반드시 발표 할때 말을 좀 느리게 해보자고 다짐했습니다. 저번에 교수님께서 지적하시기도 하셨고, 요즘따라 보이는게 엄청나게 중요하다는걸 느끼고 있기도 합니다. 무엇보다 보이는게 좋다면, 타인에게도 긍정적인 영향을 줄 수 있겠구나 하고 생각이 들었습니다. 고급 인재들이 귀한 시간 내서 저의 발표를 듣는데.. 솔직히 그 시간을 낭비하는 시간으로 만들까봐 너무 걱정되죠.

또한.. 개인연구 발표 자료를 영어 위주로 준비할지, 한글 위주로 준비할지 많은 고민을 하다가.. 제가 해외에 나가는걸 그다지 좋아하지 않기도 하고.. 앞으로 나갈 생각도 전혀 없고.. 제가 한국을 너무 좋아해서, 그냥 나중에는 한국사람을 자료를 만들면 좋겠다 싶어 한글 위주로 장표를 준비를 했습니다.

이번 개인연구 발표 제목을 이렇게 한번 정해봤습니다. “Finding the quickest way to solve pattern quiz”. 사실 이 제목은 2주 전에 정해진 제목인데, 2주전이면 재윤형이 개인연구 발표를 하던 때입니다. 그때 교수님께서 논문을 빨리 써야한다고 재윤형한테 조언을 해주셨는데, 그 조언을 옆에서 듣고.. 저도 졸업 논문에 대해 많이 불안해 졌습니다. 빨리 졸업논문을 무엇으로 쓸 지 정해야 연구를 하든 말든 할텐데.. 그런게 없어서 약간 멘붕..? 비슷한게 왔었습니다. 2주전에 일단 손에 잡히는 주제만으로라도 논문을 써보자 하는 심정으로 고른 주제가 이것이고, 와중에 제가 미약하게나마 기여했던 금감원 프로젝트를 교수님과 선배님들이 저의 국내논문으로 교수님이 양보해주셔서.. 아주아주 감사하고 행복한 심정이며 반드시 은혜를 값겠다는 포부를 이자리에서 밝힙니다.

우선 오늘 발표할 주제는 제가 2주 전에 지푸라기라도 잡는 심정으로 정했던 연구 논문 주제입니다. 미리 말씀드리자면, 이 주제는 충분히 교수님에게 거절당할 준비가 되어 있으며, 얼마든지 최후순위로 밀려날 준비가 되어있는 그런 주제입니다.

사실 이것 말고도.. 힙소트의 피폿 위치를 최적으로 잡기 위한 신경망 알고리즘 등등.. 갖가지 괴상한 주제를 생각했는데.. 들으면 아시다 싶이 정말 말도 안되는 주제입니다.. 그만큼 제가 주제를 정하는 센스랄까 그런게 많이 부족해서 논문 주제 선정에 많은 고통을 받고 있었습니다.

목차는 다음과 같습니다. 먼저 금융 NLP라고 되어있는 카테고리가 있네요.. 이 부분은 기록으로 남기면 안되는 컨텐츠라 이렇게 목차만 있고 장표는 없습니다. 사실 대단한 이야기는 절대 아니고.. 말해준 사람이 취해있었기도 하고 절대 딴데 말하지 말라고 해서.. 근데 또 저는 연구실 사람들에게 항상 감사해서 최대한 많은 정보를 전달드리고 싶어서.. 용기를 내서 넣어봤습니다. 그리고 금감원 프로젝트를 논문으로 쓸 때도 참고하게 될 이야기이긴 합니다.

금감원 프로젝트는 기존의 금융권에서 사용하던 정량적 지표에 텍스트데이터라는 정성적인 지표를 더하는 프로젝트였는데요, 실제로 이런 일을 하는 산업군은 금감원도, 한국은행도 아닌 다른곳이라고 생각합니다. 바로 밥줄을 걸고 금융 지표를 예측하려고 하는 수 많은 해지펀드들입니다. 이 해지펀드들이 엔지니어링 쪽으로 치우치면 우리는 이들을 퀀트라고 부릅니다. 저랑 술을 많이 마시다 만취해서 자신이 잘나가는 퀀트라고 큰소리치는 어떤 선배가 말하길.. 금융으로의 NLP 적용 수준은 학계보다 자기네들이 훨씬 앞서있지만 업계 성격상 절대 공개하지 않는다고 했습니다.

최근에 월드퀀트 설명회 관련해서 교수님께서 저희 연구실 구성원들에게 메일을 보낸적이 있습니다. 다른 해지펀드에 있는 사람이 말하길, 월드퀀트는 지금 약간 망해가고 있다고 합니다. 해지펀드 회사에서는 퀀트라고 불리는 수익을 얻기위한 모델을 짜는 핵심 인력들이 있는데요. 기존에 월드퀀트라는 회사는 퀀트들과 연봉 계약을 할 때 근무일 하루당 1개의 모델을 짜는것이 계약조건이라고 합니다. 그러다보니 얼토당토 않거나 과거에 오버피팅 된 모델을 만드는 경우가 많다고 합니다. 결국 요즘에 월드퀀트 보다 일을 덜 시키면서 더 높은 연봉을 주는 작은 규모의 해지펀드로 퀀트들이 이직하고 있다고 합니다. 2년전에 월드퀀트 한국 지사에서 일하던 퀀트가 20명이었다면 지금은 다 이직하고 2명만 남아있는 상태라서 곧 월드퀀트의 공격적 리쿠르팅이 시작될 것이라고 합니다. 근데 또 요즘은 카카오나 네이버도 공격적으로 리쿠르팅을 하고있어서.. 선배님들 모두 행복하게 원하는 직장에서 일하십시오!

이 말을 해준 사람은 월드퀀트에서 나와 암호화폐 해지펀드로 옮긴 선배가 해준 이야기이구요. 저는 이 암호화폐 해지펀드에서 바쁜 메인모델러가 하기 힘든 잡스러운 개발을 보수를 받고 한번 맡아준 적이 있는데요. 그냥.. 꼼꼼하게 개발해서 보내줬는데 그 꼼꼼함이 마음에 들었는지 갑자기 채용전환 이야기를 하면서 술을 사준다고 했습니다. 이때 술을 마시다가 월드퀀트 욕이랑 해지펀드에서 사용하는 NLP에 대해 말이 잠깐 나왔었는데.. 이때 들은 이야기를 한번 금감원 프로젝트에 더해서 적용해보고 싶어졌습니다.

사실 이말을 한 이유는.. 금감원관련 논문도 있지만.. 뛰어난 알고리즘을 짜지 않고 그냥 시키는거 잘 이해하고 깔끔하게만 짰는데도 엄청 좋아하더라는 신기한 현상을 공유하고 싶어서 해보았습니다.

이제 본격적으로 저의 개인 연구 후보에 대해서 이야기 해보도록 하겠습니다. 간단히 말해 저의 개인연구는 이 장표의 제목과 동일합니다. "패턴 퀴즈를 풀어내는 가장 빠른 방법을 찾기".

명확한 문제의 이해를 위해 패턴 퀴즈가 무엇인지 설명합니다. 최고의 설명으로 예시만한게 없죠! 이게 바로 제가 풀고자 하는, 제가 정의한 패턴 퀴즈의 한가지 예시입니다. 시중에는 레이븐스 프로그래시브 메트릭스라고 불리는 이 문제들은 멘사와 트리플 나인에서 이용하고 있는 테스트의 모태가 되기도 하며, 한국 공공기관에서 사용하고 있는 웩슬러 지능검사와 상관계수가 0.7이상이라고 학계에서 증명된 검사입니다. 여담으로 자폐아들이 이 문제를 잘푸는 경향이 있다고 합니다...

보시다 싶이 예시로 사용된 문제는 매우 간단한 문제이며 정답은 5번이라는 것을 바로 알 수 있습니다. 이번 발표에서는 왜 이문제는 쉬운 문제이고 그렇다면 어려운문제일수록 어떤 특징이 있는지 신경망의 관점에서 저의 견해를 밝힙니다.

이 문제를 푸는일에 있어서 몇가지 특이한 부분들이 있습니다. 이를 설명하기에 앞서..! 혹시나.. 왜 이런 연구를 하고자 하는거지? 이게 산업에 무슨 기여를 할수 있으며, 학문적으로 무슨 의미가 있어서..? 라고 생각하실수도 있는 청자를 위해 왜 이런 연구를 하고자 하는지에 대한 대답은. 매끄러운 구성을 위해 문제의 특징들을 설명하고 나서 설명드립니다.

이 문제를 푸는 일은 다음과 같은 특징을 가지고 있습니다. 첫번째로 사전지식이 필요하지 않습니다. 두번째로 각 문제의 해는 통계적으로 랜덤하지 않습니다. 세번째로 각 해는 문제의 내용이 심볼화 되었을 때, 백터공간에서의 사칙연산에 의해 구해질 수 있습니다. 

이 세개의 특징들을 하나하나 설명해보도록 하겠습니다. 첫번째인 사전지식이 많이 필요하지 않다는 말은, 말 그대로 딥 러닝을 적용해 풀고자 하는 많은 문제들에 비해서 사전에 필요한 지식이 거의 없다는 뜻입니다. 예를 들어. 자연어처리의 경우 수많은 단어 코퍼스가 받쳐줘야 하고, 지엽적이든 포괄적이든 어떠한 사전 지식에 의한 문맥에 영향을 받을 수 밖에 없다는 점입니다. 그리고 그 예로.. 삼대 지x언어라고 불리는 한국어 일본어 중국어등에서 나타나는 품사나 조사등에대한 규칙들이 이 문제에서는 거의 없습니다.

다른 예시로 컴퓨터 비전 처리를 들 수 있습니다. 딥러닝을 비전에 적용하기 위해서는 방대한 양의 데이타가 필요하며, 사진속의 객체가 현실 뇌의 지각체계에서 어떤 객체로 인식되는지에 대한 무수히 많은 연결 정보가 필요합니다. 그리고 사진에서 표현되지 않는 현실 세계의 통계적인 분포를 따르는 잠재변수들에 대한 도메인 지식이 필요하며.. 상황에 따라 빛의 삼원색 등에 대한 추가적인 규칙도 알고 있어야 합니다. 하지만 패턴퀴즈를 푸는 일은 이런 정보가 필요 없습니다. 그리고 이건 저의 견해인데, 방대한 양의 데이터가 필요 없을수도 있습니다.

두번째 특징은 문제의 해가 통계적으로 Random하지 않다는 점입니다. 다음과 같은 숫자로 심볼화 된 레이븐 매트릭스 문제가 있다고 봅시다. 딱 보면 알다싶이 이 문제는 그냥 피보나치 수열 문제입니다. 이 문제를 저희가 통계적으로 접근할 필요가 있을까요? 이 문제의 정답은 랜덤하지 않습니다. 문제 자체가 확정모형에서 만들어진 문제이기 때문에 정답을 알아내는 식도 확정모형이어야합니다.

확률모형으로 만들어진 문제가 아니라는 것은 무엇을 의미할까요? 결국 이 이야기는 이 문제를 풀기위해 랜덤성을 지울만큼 많은 데이터가 많이 필요하지 않다는 것입니다. 주어진 패턴을 설명만 가능하다면 오버피팅이 허용됩니다. 오히려 오버피팅이 권장될 수도 있습니다. 우리는 이 문제에서 8개의 시퀀셜한 데이터를 이용해 9번째의 확실한 정답을 도출하면 됩니다.

마지막 특징은 백터공간에서 사칙연산에 의해 해가 구해질 수 있다는 점입니다. 이것이 무엇을 말하는 것이냐 하면, 저희가 현시대에 사용하고 있는 컴퓨터. 즉, 튜링머신같은 논리회로와 사칙연산을 계산하는 유한상태 기계에서 해가 구해질 수 있다는 점입니다.

위 피보나치수열로 이루어진 패턴 퀴즈의 정답은 다음의 간단한 파이썬 코드와 같습니다. 일단 이 문제에 대해서는 튜링머신으로 정답을 도출하는게 가능하다는걸 코드로 증명이 된 샘입니다. 이제 우리에게 남은 작업은 어떤 신경망이 이 문제를 풀 수 있는가 라는 점입니다.

단순한 두개의 셀로 이루어진 RNN구조로 이 문제의 해답으로 역전파가 가능합니다. 이 관찰 결과가 시사하는 바는 다음과 같습니다. 두개 수의 조합으로 이루어지며, 한개의 프로세스가 반복되는 패턴문제는 이 역전파를 이용한 신경망으로 모두 풀 수 있다는 점입니다. 실제로 그런지를 알아내기 위해 파이썬으로 익명의 패턴을 만들어 내는 코드를 짜고 텐서플로우로 정말 이 구조로 모두 오버피팅시에 정답을 도출하는지 실험 하고자 합니다.

이제 매끄럽게 왜 이문제를 연구하고자 하는가에 대한 질문에 답할 수 있게 되었습니다. 사실 이 패턴퀴즈를 풀어내는 신경망 구조를 탐구해보는 연구는 다른말로 이렇게 표현 가능합니다. "확정모형으로 만들어진 패턴의 확정모형을 유추하는 신경망을 만들어 내는 문제"라고 말입니다.

지금은 추상적으로 단순 9개 패턴에 대한 문제를 풀어내고 있지만, 패턴이 더 복잡해진다면? 필요한 세포수가 많은 문제가 된다면? 레이어 여러개 짜리 RNN으로 풀 수 밖에 없는 문제라면? 저는 문제가 복잡해지다 보면 이 신경망알고리즘이 이런 문제를 풀어내는 최적의 알고리즘이 되는 분기점이 있을 것이라고 생각합니다. 이 알고리즘의 응용분야가 당장은 잘 안떠오르지만 어쩌면 역공학의 일부로써 이용될 수도 있지 않을까 감히 생각해봅니다.

그 다음으로는 거대한 신경망 알고리즘이라함은 뇌를 포함한 저희 몸의 신경체계 전부를 말합니다. 여기서 저희는 블랙박스라고 취급되는 연산과정을 거치게 됩니다.

그리고 결국 내어놓는 해답은 다시 간단한 사칙연산이 되어버립니다. 그 예시로 왼쪽의 문장과 같이 생각하는 사람들 예로 들 수 있습니다. 일정 공간에 사과 한개를 놓고 한개를 더 놓았더니 두개더라. 그래서 1 더하기 1은 2이다. 이는 우리가 보기에 확률모형이 아닌 확정모형으로 보입니다.

"냉장고에 물 한병 == 사러 나가기" 라는 단순한 패턴을 가려내기 위해 너무 복잡한 연산이 필요한것이 아닌가? 저는 딥러닝을 공부하면서 항상 이 구조가 아이러니하고 재미있다고 생각했습니다.

하지만 사람 뇌에서 만들어진 3번의 확정모형처럼 보이는 모델은, 사실 확정모형이 아닌 경험이라는 데이터에 의한 확률모형인 것입니다. 한 반에 초등학생 40명을 모아놓고 39명이 1+1은 3이라고 확언할 경우 남은 한명도 1더하기 1은 3이라고 대답하게 될 확률이 생깁니다.

인지부조화는 3번이 확정모형이라는 기존의 믿음을 깨버리는 예시가 될 수 있으며 저는 개인적으로 패턴퀴즈를 신경망 알고리즘으로 푸는 것이 좀더 사람이 생각하는 방식을 연구하는데 도움이 될 것이라고 봅니다. 예전에 본 적이 있는데, 암산을 아주 잘하는 사람은 머리회전이 빠른게 아니라 그에 맞는 주판과 비슷한 아키텍쳐를 머릿속에 그리고 있다고 합니다.

저는 어쩌면 머릿속에 주판처럼, 신경망 알고리즘을 이용해서 최적의 소팅하는 방식을 찾아내거나 할수도 있겠다는 바보같은 생각을 했다가 5초만에 이건 좀 너무 나갔다.. 라고 생각하기도 했습니다. 물론 이러한 매니악한 의견은 그냥 연구실 세미나니까 한번 말해보는 것이고 저도 이런 생각이 그닥 가치가 있을거라고 생각은 안합니다.

죄송합니다. 다음으로 넘어가서! 그렇다면 패턴퀴즈를 잘 푼다는것은 어떤것일까요? 그냥 뭐든지 풀수 있는 슈퍼 신경망을 하나 두면 되는걸까요..? 아닙니다. 이걸로는 부족합니다. 우리는 보통 똑똑한 사람을 정의할 때 어떤 문제를 빠른 시간안에 푸는 것, 즉 문제를 푸는데 걸리는 시간도 중요하게 생각합니다. 이번 발표의 제목이 패턴퀴즈를 푸는 가장 빠른방법 찾기 인 이유가 바로 여기서 만들어진 제목입니다.

이 작업을 위해 저는 크게 세가지 선행기술을 꼽아보았습니다. 첫번째로는 원샷 러닝. 제가 예전에 발표하고 다른 연구실 선배님들도 또 발표하고 이미 여러번 언급된 선행기술이라 다들 잘 알고 계실겁니다. 이 문제는 8개의 데이터를 통해 마지막 9개의 데이터를 맞출 뿐이지 각 문제는 정말 처음보는 문제라는 가정이 수반되어야합니다.
