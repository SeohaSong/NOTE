```python
def get_fibos(n, s1=0, s2=1):
    fibos = [s1, s2]
    for _ in range(n-2):
        temp = s2
        s2 = s1+s2
        s1 = temp
        fibos.append(s2)
    return fibos[:n]
```

```python
n_sample = 1000


x = np.array(range(1, n_sample+1), dtype=np.float64)
y = 2**x

plt.plot(x, y)
plt.show()


x = range(1, n_sample+1)
y = get_fibos(n_sample)

x = np.array(x, dtype=np.float64).reshape(-1, 1)
y = np.array(y, dtype=np.float64).reshape(-1, 1)

plt.plot(x, y)
plt.show()
```
